```python
teams_we = ['LAS', 'PHO', 'SAS', 'SEA', 'TUL', 'MIN']
teams_ea = ['CON', 'IND', 'NYL', 'ATL', 'WAS', 'CHI']

# construct a table of the results with the tmID sorted by alphabetical order and the playoff seed (Y/N) where Y means that the teams is in the first 4 seeds
# the table is sorted by the playoff seed
def construct_table():
    table = []
    for index, team in enumerate(teams_we):
        table.append((team, 'Y' if index < 4 else 'N'))
    for index, team in  enumerate(teams_ea):
        table.append((team, 'Y' if index < 4 else 'N'))
    table.sort(key=lambda x: x[0])
    return table

# print the table
def print_table(table):
    print('tmID,playoff')
    for team in table:
        print(team[0] + ',' + team[1])


table = construct_table()

print_table(table)

```
